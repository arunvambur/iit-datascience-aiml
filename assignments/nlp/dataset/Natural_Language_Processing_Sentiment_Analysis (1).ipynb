```python
import pandas as pd
```

```python
data = pd.read_csv('twitter_training.csv')
```

```python
data.columns = ['ID', 'Brand', 'Sentiment', 'Phrase']
```

```python
data = data.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Sentiment'] = le.fit_transform(data['Sentiment'])
```

```python
le.classes_
```

```python
data.head()
```

```python
#Feature Generation Using Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
```

```python
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True, stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts= cv.fit_transform(data['Phrase'])
```

```python
#splitting in training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_counts, data['Sentiment'], test_size = 0.3, random_state=29)
```

```python
#model Building
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
#predictions
pred = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
#tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
text_tf= tf.fit_transform(data['Phrase'])
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(text_tf, data['Sentiment'], test_size = 0.3, random_state=29)
```

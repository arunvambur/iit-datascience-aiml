### Covariance Matrix

```python
cov_mat = np.cov(X_std, rowvar=False)
cov_mat
```

```python
cov_mat.shape
```

### Compute Eigen Values and Eigen Vectors

```python
print('Covariance matrix \n')

eigen_values, eigen_vectors = np.linalg.eig(cov_mat)
print('Eigen values:\n', eigen_values)
print('Eigen vectors:\n', eigen_vectors)
```

### selecting the 'K' Eigen values/Principal Components

```python
tot = sum(eigen_values) #all of it
tot
```

```python
exp_var = [(i/tot)*100 for i in sorted(eigen_values, reverse=True)]
print('Variance Explained:', exp_var)
```

```python
eigen_values/tot*100
```

```python
73 + 23
```

#### Explained variance shows the contribution of each eigen value in explaining the variance within data. Let us say if `K=2` : two eigen values can explain 96% of variance then we can go with 2PCs

```python
cum_exp_var = np.cumsum(exp_var)
```

```python
print('Cumulative variance explained', cum_exp_var)
```

### Make a list of (eigen values, eigen vectors)

```python
eig_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]
eig_pairs
```

### Select the first two

```python
eig_pairs[0][1]
```

```python
eig_pairs[1][1]
```

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(4, 1), eig_pairs[1][1].reshape(4, 1)))
matrix_w
```

```python
Y = X_std.dot(matrix_w)
```
